```python
for scale in [10]:
    # 加载不同规模的无人机、任务和目标数据
    t_uavs, t_tasks, t_targets = load_different_scale_csv(test_uav_csv, test_task_csv, scale)
    env = UAVEnv(t_uavs, t_targets, t_tasks)

    # 运行随机算法
    start_time = time.time()
    random_allocation(env)
    elapsed = time.time() - start_time
    log_time(elapsed)

    # 运行 RR 轮询算法
    start_time = time.time()
    round_robin_allocation(env)
    elapsed = time.time() - start_time
    log_time(elapsed)

    # GA
    start_time = time.time()
    ga_scheduler = GeneticAlgorithmScheduler(env)
    ga_scheduler.run()
    elapsed = time.time() - start_time
    log_time(elapsed)

    # PSO
    start_time = time.time()
    pso = PSO(env)
    pso.optimize()
    elapsed = time.time() - start_time
    log_time(elapsed)
    
    # MOPSO
    start_time = time.time()
    mopso = HS_MOPSO(env)
    mopso.run()
    elapsed = time.time() - start_time
    log_time(elapsed)

    # DRL
    start_time = time.time()
    test_dqn(env, scale, "dqn")
    elapsed = time.time() - start_time
    log_time(elapsed)

    # GMP_DRL
    start_time = time.time()
    test_dqn(env, scale, "gmpdqn")
    elapsed = time.time() - start_time
    log_time(elapsed)

    log_n()
```
